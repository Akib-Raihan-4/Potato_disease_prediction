```python
import cv2
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('potatoes.h5')
```

```python
class_name = ['Early blight', 'Late blight', 'Healthy']
```

```python
def preprocess_frame(frame):
    # Resize the frame to match the input size of your model
    resized_frame = cv2.resize(frame, (256, 256))
    # Preprocess the image (e.g., normalize pixel values, etc.)
    preprocessed_frame = (resized_frame / 255.0).astype(np.float32)
    # Expand dimensions to create a batch of size 1
    return np.expand_dims(preprocessed_frame, axis=0)
```

```python
def get_prediction(frame):
    preprocessed_frame = preprocess_frame(frame)
    predictions = model.predict(preprocessed_frame)
    # Process predictions and get the class name and confidence
    predicted_class = class_name[np.argmax(predictions[0])]

    confidence = np.max(predictions[0]) * 100
    return predicted_class, confidence, np.argmax(predictions[0])
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Get prediction for the current frame
    try:
        predicted_class, confidence, variable = get_prediction(frame)
        print(variable)
    except Exception as e:
        print("Error during prediction:", e)
        break

    # Overlay the prediction on the frame
    cv2.putText(frame, f"Prediction: {predicted_class}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.putText(frame, f"Confidence: {confidence:.2f}%", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Potato Plant Disease Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close the window
cap.release()
cv2.destroyAllWindows()
```
